# Visualisieren von Daten

## matplotlib - Eine externe Bibliothek zur Erzeugung von Diagrammen

```python
import matplotlib.pyplot as plt  #

plt.figure(num=1, figsize=[10.0,4.8]) #Größe des Diagrams in inches 

plt.plot(1, 4, 'r+')   # Zahlenpaar x=1, y=4 mit rotem Plus plotten
plt.plot(2, 2, 'bo')   # Zahlenpaar x=2, y=2 mit blauem Kreis plotten
plt.plot(3, 7, 'g*')   # Zahlenpaar x=3, y=7 mit grünem Stern plotten
plt.plot(4, 5, 'yx')   # Zahlenpaar x=4, y=5 mit gelbem x plotten
plt.xlabel('some numbers')  #Beschriftung x-Achse
plt.ylabel('other numbers') #Beschriftung y-Achse
plt.show()                  #Plot anzeigen
```

## Plotly - Eine weitere externe Bibliothek zur Erzeugung von 2D- und 3D Graphen

Zur Visualisierung eines Datensatzes muss dieser zunächst wieder mit der Pandas-Bibliothek eingelesen werden.

```python
import pandas as pd

# Einlesen der CSV-Datei
df = pd.read_csv("temperatures.csv")
df_sorted = df.sort_values(by='Datum')

# Filtern nach Stadt "Berlin"
df_berlin = df_sorted[df_sorted['Anlage'] == 'Machine 1']
```

Jetzt verwenden wir Plotly Express, um ein 2D-Liniendiagramm zu erstellen, welches den Temperaturverlauf in einer Anlage über die Zeit darstellt.

```python
import plotly.express as px

# Erstellen des Liniendiagramms für Berlin
fig = px.line(df_berlin, x='Datum', y='Temperatur', title='Temperaturverlauf Anlage 1')
fig.show()
```

Um komplexere Zusammenhänge unter verschiedenen Daten herausfinden zu können, reicht ein 2D Graph unter Umständen nicht aus. Hier bietet sich das Hinzufügen zusätzlicher Dimensionen bei der Visualisierung an. Wir wollen also die Datenpunkte in einem 3D-Scatterplot darstellen, wobei die X-Achse das Datum, die Y-Achse die Temperatur und die Z-Achse die Luftfeuchtigkeit darstellt.

```python
# 3D-Scatterplot erstellen
fig = px.scatter_3d(df_berlin, x='Datum', y='Temperatur', z='Luftfeuchtigkeit', color='Stadt', title='Temperatur und Luftfeuchtigkeit')
fig.show()
```

```python
import plotly.express as px

# 3D-Scatterplot mit Farbe nach Stadt und Schattierung nach Temperatur
fig = px.scatter_3d(df_sorted, 
                    x='Datum', 
                    y='Temperatur', 
                    z='Luftfeuchtigkeit', 
                    color='Stadt',  # Farben je nach Stadt
                    size='Temperatur',  # Größere Punkte bei höheren Temperaturen
                    opacity=0.7,  # Transparenz
                    color_continuous_scale='Viridis',  # Farbskala für Temperatur
                    title='Temperatur, Luftfeuchtigkeit und Zeit in verschiedenen Städten')

# Diagramm anzeigen
fig.show()
```

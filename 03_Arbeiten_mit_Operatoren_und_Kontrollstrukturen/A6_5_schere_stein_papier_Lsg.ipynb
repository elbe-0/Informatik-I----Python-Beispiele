```python
import random
# Variablen deklarieren
pcWahl = 0
spielerWahl = 0

# Zufallszahl zwischen 1 und 5 generieren
random.seed()
pcWahl = random.randint(1,5)  # Zufallszahl von 1 bis inklusive 5

# Eingabeaufforderung Spieler
print("Bite wählen Sie eine Figur:")
print("1 - Schere")
print("2 - Papier")
print("3 - Stein")
print("4 - Echse")
print("5 - Spock")
spielerWahl = int(input())  # Wichtig: den Input als Integer speichern

# Regeln anwenden
print("Der PC hat", pcWahl, " gewählt")

# 1. Fall: PC hat Schere gewählt
if (pcWahl == 1):
    if(spielerWahl == 1):
        print("unentschieden")
    elif(spielerWahl == 2):
        print("Schere schneidet Papier - verloren")
    elif(spielerWahl == 3):
        print("Stein zerstört Schere - gewonnen")
    elif(spielerWahl == 4):
        print("Schere köpft Echse - verloren")
    elif(spielerWahl == 5):
        print("Spock zertrümmert Schere - gewonnen")
# 2. Fall: PC hat Papier gewählt
if (pcWahl == 2):
    if(spielerWahl == 1):
        print("Schere schneidet Papier - gewonnen")
    elif(spielerWahl == 2):
        print("unentschieden")
    elif(spielerWahl == 3):
        print("Papier wickelt Stein ein - verloren")
    elif(spielerWahl == 4):
        print("Echse frisst Papier - gewonnen")
    elif(spielerWahl == 5):
        print("Papier widerlegt Spock - verloren")
# 3. Fall: PC hat Stein gewählt
if (pcWahl == 3):
    if(spielerWahl == 1):
        print("Stein zerstört Schere - verloren")
    elif(spielerWahl == 2):
        print("Papier wickelt Stein ein - gewonnen")
    elif(spielerWahl == 3):
        print("unentschieden")
    elif(spielerWahl == 4):
        print("Stein zerquetscht Echse - verloren")
    elif(spielerWahl == 5):
        print("Spock verdampft Stein - gewonnen")
# 4. Fall: PC hat Echse gewählt
if (pcWahl == 4):
    if(spielerWahl == 1):
        print("Schere köpft Echse - gewonnen")
    elif(spielerWahl == 2):
        print("Echse frisst Papier - verloren")
    elif(spielerWahl == 3):
        print("Stein zerquetscht Echse - gewonnen")
    elif(spielerWahl == 4):
        print("unentschieden")
    elif(spielerWahl == 5):
        print("Echse vergiftet Spock - verloren")       
# 5. Fall: PC hat Spock gewählt
if (pcWahl == 5):
    if(spielerWahl == 1):
        print("Spock zertrümmert Schere - verloren")
    elif(spielerWahl == 2):
        print("Papier widerlegt Spock - gewonnen")
    elif(spielerWahl == 3):
        print("Spock verdampft Stein - verloren")
    elif(spielerWahl == 4):
        print("Echse vergiftet Spock - gewonnen")
    elif(spielerWahl == 5):
        print("unentschieden")         
```
